```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Cost_Overrun.csv')
```

```python
df.head(10)
```

```python
X=df.iloc[:,4:52]
y=df.iloc[:,2:3]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
clf=RandomForestRegressor(n_estimators=100, random_state=1)
clf.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
model_ranks=pd.Series (clf.feature_importances_,index=X_train.columns, name="Importance").sort_values (ascending=True, inplace=False)
fig=plt.figure(figsize=(15,10))
ax=model_ranks.plot(kind='barh')
plt.savefig("fig.png")
plt.xlabel("Risk Level")
plt.ylabel("Causes")
```

```python
from sklearn.metrics import classification_report, accuracy_score
predict = clf.predict(X_test)
accuracy = accuracy_score(y_test,predict.round())
print(f"Accuracy is {round(accuracy*100,2)}%")
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print('Model_Name:- RFGA')
pred = clf.predict(X_test)
cnfm = pd.DataFrame(confusion_matrix(y_test,pred.round()))
clsf = pd.DataFrame(classification_report(y_test,pred.round(),output_dict=True))
cnfm
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(cnfm, annot=True, cmap='Greens', fmt='d',annot_kws={'rotation': 'horizontal'})
plt.title('Confusion Matrix')
plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')
plt.show()
```
